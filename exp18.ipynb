```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [0, 70, 30, 90, 60],
    'Sales': np.random.randint(100, 500, 5),
    'Profit': np.random.randint(10, 100, 5)

})
display(df)


#1. Area plot Graph Code

plt.fill_between(df['Category'], df['Value'], color='green', alpha=0.5)
plt.title('Bhomik made this Area plot graph')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


#2.Sales and profit area plot graph

sns.set_style("whitegrid") # apply seaborn style
plt.figure(figsize=(7,4))
plt.fill_between(df['Category'], df['Sales'],
                 color='skyblue',alpha=0.5, label='Sales')
plt.fill_between(df['Category'], df['Profit'],
                  color='orange',alpha=0.5, label='Profit')
plt.title("Bhomik Keshi made this Sales and Profit Area Plot Graph")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#3. Pie Chart

plt.figure()
plt.pie(df['Value'], labels=df['Category'], autopct='%1.2f%%')
plt.title("Bhomik Keshi made this Pie Chart")
plt.show()
print()
print()
```

```python
#3. Donut Chart
plt.figure()
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Bhomik made this donut graph")
plt.show()
```

```python
# 4. box plot (Outliners Detection)

plt.figure()
sns.boxplot(x=df['Value'])
plt.title("Bhomik Keshi made this Box Plot")
plt.show()
```

```python
#5. Heatmap

plt.figure()
corr = df[['Sales','Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Bhomik made this Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# select only numeric columns

corr = df[['Value','Sales','Profit']].corr()
plt.figure(figsize=(5,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Bhomik made this Heatmap')
plt.xlabel('Category')
plt.ylabel('Proftit')
plt.show()
```

```python
#Bubble Plot (3 Variable Chart)
import matplotlib.pyplot as plt
plt.scatter(df['Sales'], df['Profit'],
            s=df['Value']*10)           #size of bubbles
plt.title("Bhomik made this Bubble Plot: Sales vs Profit by Values")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales',y='Profit',size='Value',hue='Value',data=df,sizes=(50,300),palette="viridis")
plt.title('bhomik')
plt.show()
```

```python


#2. Second Dataset
np.random.seed(7)
n = 100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df['Loan_Status'] = np.where(df['Credit_Score'] > 600, 'Approved', 'Rejected')

print(df.head())
```

```python

#Boxplot (Outlier Detection – Loan Amount)
import matplotlib.pyplot as plt

plt.boxplot(df['Loan_Amount'])

plt.title("Loan Amount Distribution (Boxplot)")
plt.ylabel("Loan Amount")

plt.show()
```

```python

#3. Histogram (Credit Score Distribution)
plt.hist(df['Credit_Score'], bins=6, color='skyblue', edgecolor='black')

plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")

plt.show()
```

```python

#Bubble Plot (Advanced Scatter)
plt.scatter(df['Income'], df['Loan_Amount'],
            s=df['Credit_Score']/2,
            c=df['Credit_Score'],
            cmap='viridis',
            alpha=0.6)

plt.colorbar(label='Credit Score')

plt.title("Bubble Plot (Income vs Loan)")
plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.show()
```

```python

#6. Correlation Heatmap (Very Important)
import seaborn as sns

corr = df[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
